## Overview
This notebook provides an independent analysis of Nike's fiscal year 2024 financial performance.  
All data is sourced directly from Nike's publicly available **10-K annual report** filed with the SEC.

We examine:
- Revenue by geographic segment (North America, EMEA, Greater China, APLA)
- Revenue by product category (Footwear, Apparel, Equipment)
- Distribution channel performance (Wholesale vs. Direct-to-Consumer)
- Year-over-year comparisons (FY2023 vs FY2024)
- Key findings and strategic recommendations


## 1. Import Libraries & Load Data

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

print("Libraries loaded successfully.")
print(f"Pandas version: {pd.__version__}")
```

## 2. Data Preparation

All figures are in **millions of USD** as reported in Nike's FY2024 10-K filing.  
Source: Nike, Inc. Form 10-K — Note 14: Revenues Disaggregation (SEC EDGAR)


```python
# ── REGIONAL REVENUE ──────────────────────────────────────────────
regional_data = {
    'Region': ['North America', 'EMEA', 'Greater China', 'APLA'],
    'FY2023': [21608, 13418, 7248, 6431],
    'FY2024': [21396, 13607, 7545, 6729]
}
df_regional = pd.DataFrame(regional_data)
df_regional['YoY_Change'] = df_regional['FY2024'] - df_regional['FY2023']
df_regional['YoY_Pct'] = ((df_regional['FY2024'] - df_regional['FY2023']) / df_regional['FY2023'] * 100).round(2)
df_regional['FY2024_Share'] = (df_regional['FY2024'] / df_regional['FY2024'].sum() * 100).round(1)

# ── PRODUCT CATEGORY REVENUE ──────────────────────────────────────
product_data = {
    'Category': ['Footwear', 'Apparel', 'Equipment', 'Other'],
    'FY2023': [35290, 13933, 1755, 239],
    'FY2024': [35227, 13868, 2112, 155]
}
df_product = pd.DataFrame(product_data)
df_product['YoY_Change'] = df_product['FY2024'] - df_product['FY2023']
df_product['YoY_Pct'] = ((df_product['FY2024'] - df_product['FY2023']) / df_product['FY2023'] * 100).round(2)
df_product['FY2024_Share'] = (df_product['FY2024'] / df_product['FY2024'].sum() * 100).round(1)

# ── CHANNEL REVENUE ───────────────────────────────────────────────
channel_data = {
    'Channel': ['Wholesale', 'Direct-to-Consumer'],
    'FY2023': [28696, 22282],
    'FY2024': [28856, 22351]
}
df_channel = pd.DataFrame(channel_data)
df_channel['YoY_Change'] = df_channel['FY2024'] - df_channel['FY2023']
df_channel['YoY_Pct'] = ((df_channel['FY2024'] - df_channel['FY2023']) / df_channel['FY2023'] * 100).round(2)

# ── CHANNEL BY REGION (FY2024) ────────────────────────────────────
channel_region_data = {
    'Region': ['North America', 'EMEA', 'Greater China', 'APLA'],
    'Wholesale': [11004, 8562, 4262, 3930],
    'DTC': [10392, 5045, 3283, 2799]
}
df_channel_region = pd.DataFrame(channel_region_data)

print("✓ Data loaded successfully")
print(f"\n{'─'*50}")
print(f"  Total Nike Revenue FY2024: ${51362:,}M (${51362/1000:.1f}B)")
print(f"  Total Nike Revenue FY2023: ${51217:,}M (${51217/1000:.1f}B)")
print(f"  YoY Change: +${145:,}M (+0.3%)")
print(f"{'─'*50}")
```

## 3. Key Performance Indicators

```python
# Summary KPI table
kpi_data = {
    'Metric': [
        'Total Revenue',
        'Footwear Revenue',
        'Apparel Revenue', 
        'North America Revenue',
        'EMEA Revenue',
        'Greater China Revenue',
        'APLA Revenue',
        'Direct-to-Consumer Revenue',
        'Wholesale Revenue'
    ],
    'FY2024 ($M)': [51362, 35227, 13868, 21396, 13607, 7545, 6729, 22351, 28856],
    'FY2023 ($M)': [51217, 35290, 13933, 21608, 13418, 7248, 6431, 22282, 28696],
    'YoY Change ($M)': [145, -63, -65, -212, 189, 297, 298, 69, 160],
    'YoY Change (%)': ['+0.3%', '-0.2%', '-0.5%', '-1.0%', '+1.4%', '+4.1%', '+4.6%', '+0.3%', '+0.6%']
}

df_kpi = pd.DataFrame(kpi_data)
df_kpi = df_kpi.set_index('Metric')
print("Nike FY2024 Key Performance Indicators")
print("=" * 65)
print(df_kpi.to_string())
print("=" * 65)
print("Source: Nike, Inc. 10-K FY2024 — SEC EDGAR")
```

## 4. Regional Revenue Analysis

Nike reports revenue across four geographic segments.  
North America is the largest but the only segment that declined YoY.


```python
print("Regional Revenue Summary (FY2023 vs FY2024)")
print("=" * 60)
for _, row in df_regional.iterrows():
    direction = "▲" if row['YoY_Pct'] > 0 else "▼"
    print(f"{row['Region']:<20} FY2024: ${row['FY2024']:,}M  |  "
          f"YoY: {direction} {abs(row['YoY_Pct'])}%  |  "
          f"Share: {row['FY2024_Share']}%")
print("=" * 60)

# Regional bar chart — FY2023 vs FY2024
fig_region = go.Figure()

fig_region.add_trace(go.Bar(
    name='FY2023',
    x=df_regional['Region'],
    y=df_regional['FY2023'],
    marker_color='#d1d5db',
    text=[f'${v/1000:.1f}B' for v in df_regional['FY2023']],
    textposition='outside',
))

fig_region.add_trace(go.Bar(
    name='FY2024',
    x=df_regional['Region'],
    y=df_regional['FY2024'],
    marker_color='#f5a623',
    text=[f'${v/1000:.1f}B' for v in df_regional['FY2024']],
    textposition='outside',
))

fig_region.update_layout(
    title='Nike Brand Revenue by Region: FY2023 vs FY2024 ($ Millions)',
    barmode='group',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    yaxis=dict(title='Revenue ($ Millions)', gridcolor='#f3f4f6'),
    xaxis=dict(title='Geographic Segment'),
    height=450
)

fig_region.show()
```

```python
# Regional share donut chart
fig_donut_region = go.Figure(go.Pie(
    labels=df_regional['Region'],
    values=df_regional['FY2024'],
    hole=0.5,
    marker_colors=['#f5a623', '#1a1a1a', '#d97706', '#9ca3af'],
    textinfo='label+percent',
    hovertemplate='%{label}<br>Revenue: $%{value:,}M<br>Share: %{percent}<extra></extra>'
))

fig_donut_region.update_layout(
    title='Nike Brand Revenue Share by Region — FY2024',
    annotations=[dict(text='$49.3B<br>Nike Brand', x=0.5, y=0.5, font_size=13, showarrow=False)],
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    height=420
)

fig_donut_region.show()
```

## 5. Product Category Analysis

Footwear accounts for nearly 69% of Nike's total revenue —  
a significant concentration that creates strategic risk if sneaker trends shift.


```python
print("Product Category Revenue Summary (FY2023 vs FY2024)")
print("=" * 60)
for _, row in df_product.iterrows():
    direction = "▲" if row['YoY_Pct'] > 0 else "▼"
    print(f"{row['Category']:<12} FY2024: ${row['FY2024']:,}M  |  "
          f"YoY: {direction} {abs(row['YoY_Pct'])}%  |  "
          f"Share: {row['FY2024_Share']}%")
print("=" * 60)

# Product bar chart
fig_product_bar = go.Figure()

fig_product_bar.add_trace(go.Bar(
    name='FY2023',
    x=df_product['Category'],
    y=df_product['FY2023'],
    marker_color='#d1d5db',
    text=[f'${v/1000:.1f}B' for v in df_product['FY2023']],
    textposition='outside',
))

fig_product_bar.add_trace(go.Bar(
    name='FY2024',
    x=df_product['Category'],
    y=df_product['FY2024'],
    marker_color='#f5a623',
    text=[f'${v/1000:.1f}B' for v in df_product['FY2024']],
    textposition='outside',
))

fig_product_bar.update_layout(
    title='Nike Revenue by Product Category: FY2023 vs FY2024 ($ Millions)',
    barmode='group',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    yaxis=dict(title='Revenue ($ Millions)', gridcolor='#f3f4f6'),
    xaxis=dict(title='Product Category'),
    height=420
)

fig_product_bar.show()
```

## 6. Distribution Channel Analysis

Nike has been strategically shifting from wholesale to Direct-to-Consumer (DTC).  
DTC now represents **43.5%** of Nike Brand revenue — up from ~30% five years ago.


```python
print("Channel Revenue Summary (FY2023 vs FY2024)")
print("=" * 55)
for _, row in df_channel.iterrows():
    direction = "▲" if row['YoY_Pct'] > 0 else "▼"
    print(f"{row['Channel']:<25} FY2024: ${row['FY2024']:,}M  |  YoY: {direction} {abs(row['YoY_Pct'])}%")
print("=" * 55)

dtc_share = 22351 / (22351 + 28856) * 100
print(f"\nDTC as % of Nike Brand Revenue: {dtc_share:.1f}%")
print(f"Wholesale as % of Nike Brand Revenue: {100-dtc_share:.1f}%")

# Channel by region stacked bar
fig_channel = go.Figure()

fig_channel.add_trace(go.Bar(
    name='Wholesale',
    x=df_channel_region['Region'],
    y=df_channel_region['Wholesale'],
    marker_color='#1a1a1a',
    text=[f'${v/1000:.1f}B' for v in df_channel_region['Wholesale']],
    textposition='inside',
    textfont=dict(color='white')
))

fig_channel.add_trace(go.Bar(
    name='Direct-to-Consumer',
    x=df_channel_region['Region'],
    y=df_channel_region['DTC'],
    marker_color='#f5a623',
    text=[f'${v/1000:.1f}B' for v in df_channel_region['DTC']],
    textposition='inside',
    textfont=dict(color='white')
))

fig_channel.update_layout(
    title='Wholesale vs. Direct-to-Consumer by Region — FY2024 ($ Millions)',
    barmode='stack',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1),
    yaxis=dict(title='Revenue ($ Millions)', gridcolor='#f3f4f6'),
    xaxis=dict(title='Geographic Segment'),
    height=450
)

fig_channel.show()
```

## 9. Conclusion

Nike remains the world's most valuable sports brand — but FY2024 reveals a company at a critical inflection point.  

**The core narrative in three sentences:**

Revenue grew just 0.3% on a $51B base — essentially flat. North America, their largest market at 43% of revenue, declined. International markets (EMEA +1.4%, APLA +4.6%, China +4.1%) are the growth story, and the strategic shift from wholesale to DTC is accelerating.

The opportunity is clear: Nike has the brand equity, the product, and the global infrastructure to accelerate growth — but it requires deliberate investment in international markets, apparel category development, and DTC deepening in its home market.
